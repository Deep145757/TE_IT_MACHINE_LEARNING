```python
import pandas as pd
import csv
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = []
with open('/content/drive/MyDrive/Colab Notebooks/DataSet/TV Shows - Association Rule Learning.csv') as file:
  reader = csv.reader(file,delimiter=',')
  for row in reader:
    dataset += [row]
```

```python
dataset[0:1]
```

```python
len(dataset)
```

```python
te = TransactionEncoder()
x = te.fit_transform(dataset)
```

```python
x
```

```python
te.columns_
```

```python
df = pd.DataFrame(x, columns=te.columns_)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
len(te.columns_)
```

```python
df.head()
```

```python
freq_itemset = apriori(df,min_support=0.01,use_colnames=True)
```

```python
freq_itemset
```

```python
rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.25)
```

```python
rules
```

```python
rules = rules[['antecedents','consequents','support','confidence']]
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

# Iterate through the rules DataFrame
for _, row in rules.iterrows():
    antecedents = row['antecedents']
    consequents = row['consequents']
    support = row['support']

    # Add edges to the graph for each association rule
    for antecedent in antecedents:
        for consequent in consequents:
            G.add_edge(antecedent, consequent, weight=support)

plt.figure(figsize=(16, 10))

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=7000, node_color='skyblue', font_size=8, edge_color='black', width=2, alpha=0.7, arrows=True)

plt.show()
```

```python
def visualize_rules(min_confidence):
    G = nx.DiGraph()

    for _, row in rules.iterrows():
        antecedents = row['antecedents']
        consequents = row['consequents']
        support = row['support']

        for antecedent in antecedents:
            for consequent in consequents:
                G.add_edge(antecedent, consequent, weight=support)

    plt.figure(figsize=(16, 10))
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=7000, node_color='skyblue', font_size=10, edge_color='black', width=2, alpha=0.7, arrows=True)
    plt.title(f'Min Confidence: {min_confidence}')

# Call the visualization function
visualize_rules(0.2)
```
